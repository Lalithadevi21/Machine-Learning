#Auto random state

```python
sc_list = []
for i in range(1000):
  xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=i)
  lrmodel=LogisticRegression()
  lrmodel.fit(xtrain,ytrain)
  sc=lrmodel.score(xtest,ytest)
  sc_list.append(sc)

m=max(sc_list)
ind=sc_list.index(m)
print("Max score",m)
print("Random State",ind)
```
